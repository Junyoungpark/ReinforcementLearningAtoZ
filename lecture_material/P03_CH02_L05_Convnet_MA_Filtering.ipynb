```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2019, 1, 1)
end = datetime.datetime(2020, 6, 1)
```

```python
appl = web.DataReader('AAPL','yahoo',start, end)
```

```python
appl.head()
```

```python
orders = range(5,25,5)
for i in orders:
    appl['SMA_{}'.format(i)] = appl.iloc[:,0].rolling(window=i).mean()
```

```python
plt.figure(figsize=(10,5))
plt.grid(True)
plt.plot(appl['High'],label='data')
for i in orders:
    plt.plot(appl['SMA_{}'.format(i)],label='SMA {}'.format(i))
plt.legend(loc=2)
plt.title("Apple stock price", size=15)
```
