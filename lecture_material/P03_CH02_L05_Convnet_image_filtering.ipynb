```python
from scipy import ndimage, misc
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(20,10))
plt.gray()  # show the filtered result in grayscale
ax1 = fig.add_subplot(121)  # left side
ax2 = fig.add_subplot(122)  # right side
ascent = misc.ascent()
result = ndimage.sobel(ascent)
ax1.imshow(ascent)
ax2.imshow(result)
plt.show()
```

```python
import PIL.Image as pilimg

im = pilimg.open('cat.jpg').convert('RGB')
r,g,b = im.split()
```

```python
fig, axes = plt.subplots(1,4, figsize=(20, 8))
axes[0].imshow(im)
axes[0].axis('off')
axes[0].set_title("RGB Image", size=20)

axes[1].imshow(r)
axes[1].axis('off')
axes[1].set_title("R Image", size=20)

axes[2].imshow(g)
axes[2].axis('off')
axes[2].set_title("G Image", size=20)

axes[3].imshow(b)
axes[3].axis('off')
axes[3].set_title("B Image", size=20)
```
