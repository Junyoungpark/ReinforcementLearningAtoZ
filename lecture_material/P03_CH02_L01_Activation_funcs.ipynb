```python
import torch
import torch.nn as nn

import matplotlib.pyplot as plt
```

```python
acts = ['ReLU', 'LeakyReLU', 'Tanh', 'ELU', 'Sigmoid', 'PReLU', 'SELU', 'Softplus']
act_val = dict()
xs = torch.linspace(-10,10,100)
for act in acts:
    ys = getattr(nn, act)()(xs).detach()
    act_val[act] = ys


#ys = nn.ReLU()(xs)
```

```python
import numpy as np

fig, axes = plt.subplots(2, 4, figsize=(12,6))
axes = np.reshape(axes, -1)    
for ax, (act, ys) in zip(axes, act_val.items()):
    ax.grid()
    _ = ax.set_title(act, size=15)
    ax.plot(xs, ys)
fig.tight_layout()
```
